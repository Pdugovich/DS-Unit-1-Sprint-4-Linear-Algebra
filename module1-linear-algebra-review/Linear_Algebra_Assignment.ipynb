## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
# Dimensions is rows x columns
# Matricies can be multiplied together of the columns of the first match the
#rows of the second

# 3x2
# Can be multiplied by C and E
A = np.array([[1,2],
              [3,4],
              [5,6]])

# 1x3
# Can be multiplied by A and D
B = np.array([2,4,6])

# 2x3
# Can be multiplied by A and D
C= np.array([[9,6,3],
             [4,7,11]])

# 3x3
# Can be multiplied by A
D= np.array([[1,0,0],
             [0,1,0],
             [0,0,1]])

# 2x2
# Can be multiplied by C
E = np.array([[1,3],
              [5,7]])
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
CD = np.matmul(C,D)
AE = np.matmul(A,E)
BA = np.matmul(B,A)

```

```
# Matricies can be multiplied if the columns of the first are the same length
# as the rows of the second. The shape of the resulting matrix has the unused
# dimensions. So for 2x3 and 3x4, the resulting matrix would be 2x4

print(CD)
print(AE)
print(BA)
```

# Part 3 - Square Matrices

## 3.1 Find $IG$ (be sure to show your work) 😃

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
# To find the inverse of a square matrix, switch the positions of the top left
#and bottom right. Then make the other two negative. Then multiply the resulting
#matrix by 1 over the determinate of the original matrix.

#The first step of inverting G
G = np.array([[12,11],
              [7,10]])

G_1 = np.array([[10,-11],
                [-7,12]])

G_determinate = 1 / ((G[0][0] * G[1][1]) - (G[0][1]*G[1][0]))

G_I = G_determinate * G_1
```

```
G_I
```

```
#Testing against np to make sure it's right

np.linalg.inv(G)
```
