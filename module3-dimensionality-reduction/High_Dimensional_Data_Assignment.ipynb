# The Curse of Dimensionality

## 6.1 What are some of the challenges of working with high dimensional spaces?

As more dimensions are added, the distance between observations becomes more uniform, so calculating distance between points yields less and less informative results.

The computational workload increases exponentially with each higher dimension.

The effect of relevant observations is diminished with higher dimensions.

And the liklihood of overfitting the data increases with more dimensions.

```
import os.path
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import pandas as pd

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)
```

```
# Look at first 5 rows
print(national.shape)
national.head()
```

```
# Look at datatypes
# a lot of object datatypes even though they look to be strings of numbers.
national.dtypes
```

```
# check for null values
national.isnull().sum().any()
```

```
# check for number of categorical vs numeric columns
cat_cols = national.columns[national.dtypes=='object']
num_cols = national.columns[national.dtypes!='object']

print(f'{len(cat_cols)} categorical columns')
print(f'{len(num_cols)} numerical columns')
```

```
national_processed = national.copy()

# Get One-Hot-Encoding of categorical variables
national_processed = pd.get_dummies(national_processed)
print(national_processed.shape)
national_processed.head()
```

```
# Dropping the object columns per Ryan's instruction
only_num = national.drop(columns=national.columns[national.dtypes=='object'])
```

```
only_num.head()
```

```
from numpy import array
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Selecting the matrix to be standardized
X = only_num

# Use this handy sklearn function to subtract the data from the means
# then multiply them by their variance to get z-standardized data
scaler = StandardScaler()
Z = scaler.fit_transform(X)


# Use the PCA(2) to find the two longest eigenvalues
pca = PCA(2)

# fitting the pcas to the data
pca.fit(Z)
# access values and vectors
print("\n Eigenvectors: \n", pca.components_)
print("\n Eigenvalues: \n",pca.explained_variance_)

# Map onto dimension
B = pca.transform(Z)
print("\n Projected Data: \n", B)
```

```
#Converting the project values onto a dataframe to plot more easily 
df_B = pd.DataFrame(B)
```

```
fig = plt.figure(figsize= (10,10))
plt.scatter(df_B[0],df_B[1], alpha=.3)
```
